# Head to Head Model Notebook

Set `kaggle_dir` if you want the historical training pipeline to use Kaggle March Machine Learning Mania CSVs. The public Bart fetch path still works without Kaggle, but training and backtesting are stronger when those files are present.

```
from march_madness_2026_core import fit_or_load_models, get_current_team_table
```

```
kaggle_dir = None
cache_dir = 'cache'
matchup_model, score_model = fit_or_load_models(model_dir='artifacts/models', kaggle_dir=kaggle_dir, cache_dir=cache_dir, force=False)
team_table = get_current_team_table(year=2026, cache_dir=cache_dir)
matchup_model.get('metrics', {})
```

```
from march_madness_2026_core import predict_two_teams
result = predict_two_teams('Duke', 'Houston', team_table, matchup_model, score_model, recent_form_weight=1.0, program_weight=1.0, roster_weight=0.1, gear=0, upset_factor=0.0)
{k: v for k, v in result.items() if k not in {'score'}}
```
